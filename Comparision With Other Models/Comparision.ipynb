```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Setting a theme and aesthetic settings for seaborn plots
sns.set_theme(style="whitegrid")

# Data
models = ['BERT NLP', 'SVM', 'Logistic Regression', 'Random Forest']
accuracy = [96.18, 92.53, 89.06, 89.08]
precision = [0.9627, 0.9292, 0.9059, 0.88]
recall = [0.9599, 0.9205, 0.8715, 0.90]
f1_score = [0.9613, 0.9248, 0.8884, 0.89]

# Plotting
fig, ax = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Performance Comparison of Different Models', fontsize=16, y=1.05)

# Colors (highlighting BERT NLP in blue)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# Accuracy Plot
sns.barplot(x=models, y=accuracy, palette=colors, ax=ax[0, 0])
ax[0, 0].set_title('Accuracy Comparison')
ax[0, 0].set_ylim([85, 100])
ax[0, 0].set_ylabel('Accuracy (%)')

# Precision Plot
sns.barplot(x=models, y=precision, palette=colors, ax=ax[0, 1])
ax[0, 1].set_title('Precision Comparison')
ax[0, 1].set_ylim([0.8, 1])
ax[0, 1].set_ylabel('Precision')

# Recall Plot
sns.barplot(x=models, y=recall, palette=colors, ax=ax[1, 0])
ax[1, 0].set_title('Recall Comparison')
ax[1, 0].set_ylim([0.8, 1])
ax[1, 0].set_ylabel('Recall')

# F1 Score Plot
sns.barplot(x=models, y=f1_score, palette=colors, ax=ax[1, 1])
ax[1, 1].set_title('F1 Score Comparison')
ax[1, 1].set_ylim([0.8, 1])
ax[1, 1].set_ylabel('F1 Score')

# Layout adjustments
plt.tight_layout()
plt.show()
```
